```python
def detect_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)]
    return outliers

# Apply outlier detection to all numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns
outlier_counts = {col: len(detect_outliers_iqr(df, col)) for col in numeric_columns}

for col, count in outlier_counts.items():
    print(f"Column {col} contains {count} outlier values.")

df_cleaned = df.copy()
for col in numeric_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

print("\nOutliers removed, dataset size after cleaning:", df_cleaned.shape)
```

# Data Analysis

**1. What is the most selling category?**

```python
df_cleaned.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

**2. What are the most profitable states?**

```python
df_cleaned.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)
```

**3. Is there a relationship between discount and profit?**

```python
sns.scatterplot(data=df_cleaned, x='Discount', y='Profit')
plt.title('Discount vs Profit')
plt.show()
```

**4. What are your best-selling products in terms of quantity?**

```python
df_cleaned.groupby('Product.Name')['Quantity'].sum().sort_values(ascending=False).head(10)
```

**5. How are sales distributed across markets?**

```python
df_cleaned.groupby('Market')['Sales'].sum().sort_values(ascending=False)
```

**6. Does Order Priority affect profitability?**

```python
df_cleaned.groupby('Order.Priority')['Profit'].mean().sort_values(ascending=False)
```

**7. What do sales look like across months (using weeknum or Order.Date)?**

```python
df_cleaned['Order.Date'] = pd.to_datetime(df_cleaned['Order.Date'])
df_cleaned['Month'] = df_cleaned['Order.Date'].dt.month

df_cleaned.groupby('Month')['Sales'].sum().plot(kind='bar', title='Monthly Sales')
```

**8. What is the impact of Ship Mode on shipping cost?**

```python
df_cleaned.groupby('Ship.Mode')['Shipping.Cost'].mean().sort_values(ascending=False)
```

**9. What is the most profitable subcategory?**

```python
df_cleaned.groupby('Sub.Category')['Profit'].sum().sort_values(ascending=False).head(10)
```

**10. What is the distribution of the three segments in terms of total sales?**

```python
df_cleaned.groupby('Segment')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', title='Sales by Segment')
```
