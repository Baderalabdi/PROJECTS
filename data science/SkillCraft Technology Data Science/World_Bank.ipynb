# Data Analysis


```python
year = "2020"
indicator = "Population, total"

data_bar = df[df["Indicator Name"] == indicator][["Country Name", year]]

data_bar = data_bar.dropna().sort_values(by=year, ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(data_bar["Country Name"], data_bar[year])
plt.xticks(rotation=45)
plt.title(f"Top 10 Countries by {indicator} in {year}")
plt.ylabel(indicator)
plt.show()
```

```python
data_hist = df[df["Indicator Name"] == indicator][year].dropna()

plt.figure(figsize=(8,5))
plt.hist(data_hist, bins=20, edgecolor="black")
plt.title(f"Distribution of {indicator} across countries in {year}")
plt.xlabel(indicator)
plt.ylabel("Frequency")
plt.show()
```
