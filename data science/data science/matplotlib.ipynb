```python

#from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(projection='3d')
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)


ax.plot_surface(x, y, z , cmap = "hot" , edgecolor = "w")
ax.set_zlim(-0.5 ,0.5)
ax.view_init(-50,26)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```

```python
import matplotlib.patches as pat 
import matplotlib.pyplot as plt

c = pat.Circle((0.5,0.5),radius = 0.3,
        edgecolor = "red",facecolor = "blue" , alpha = 0.8)
ax = plt.axes()
ax.set_xlim(-1,2)
ax.set_ylim(-0.5,1.5)
ax.add_patch(c)
```
